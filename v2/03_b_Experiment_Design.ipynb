> - **Entrenamiento**: Modelo entrenado en datos sintÃ©ticos

### SynthCity Benchmarks
> MÃ©tricas recomendadas:
> - **C-index**: Concordancia de rankings de riesgo
> - **IBS (Integrated Brier Score)**: CalibraciÃ³n probabilÃ­stica

## Protocolo Experimental TSTR

### Principios Fundamentales

1. **El Test Set Real es SAGRADO e INMUTABLE**
   - Nunca se entrena con datos de test
   - Nunca se generan sintÃ©ticos basados en test
   - Solo se usa para evaluaciÃ³n final

2. **Semilla Aleatoria Fija**
   - `RANDOM_STATE = 42` para todos los experimentos
   - Garantiza reproducibilidad

3. **ValidaciÃ³n SIEMPRE en Datos Reales**
   - Aunque se entrene con sintÃ©ticos, la evaluaciÃ³n es en test real

---

## Matriz de Experimentos

| Escenario | Train Data | Eval Data | PropÃ³sito |
|-----------|------------|-----------|-----------|
| **Baseline** | Real (N=296) | Real Test | LÃ­nea base |
| **Augmented x1** | Real + 100% Synthetic (Nâ‰ˆ600) | Real Test | Aumento moderado |
| **Augmented x2** | Real + 200% Synthetic (Nâ‰ˆ900) | Real Test | Aumento agresivo |
| **Synthetic Only** | 100% Synthetic (N=296) | Real Test | DiagnÃ³stico de privacidad |

---

### Detalle de Escenarios

#### ğŸ”µ Baseline (Real Only)
```
Train: 296 registros reales
Test:  75 registros reales (20% holdout)
```
**PropÃ³sito**: Establecer la lÃ­nea base de rendimiento

#### ğŸŸ¢ Augmented x1 (Real + 100% Synthetic)
```
Train: 296 reales + 296 sintÃ©ticos = 592 total
Test:  75 registros reales
```
**PropÃ³sito**: Evaluar si duplicar datos mejora el modelo

#### ğŸŸ¡ Augmented x2 (Real + 200% Synthetic)
```
Train: 296 reales + 592 sintÃ©ticos = 888 total
Test:  75 registros reales
```
**PropÃ³sito**: Evaluar beneficio marginal de mÃ¡s datos

#### ğŸ”´ Synthetic Only (DiagnÃ³stico)
```
Train: 296 registros sintÃ©ticos (sin reales)
Test:  75 registros reales
```
**PropÃ³sito**: Evaluar si el sintetizador captura la distribuciÃ³n real

---

## MÃ©tricas de Ã‰xito

### MÃ©tricas Primarias

| MÃ©trica | FÃ³rmula | InterpretaciÃ³n | Umbral de Ã‰xito |
|---------|---------|----------------|-----------------|
| **C-index** | $\frac{\text{concordant pairs}}{\text{comparable pairs}}$ | Capacidad de ranking | > 0.55 |
| **AUC** (para clasificaciÃ³n) | Ãrea bajo curva ROC | DiscriminaciÃ³n | > 0.60 |

### MÃ©tricas Secundarias

| MÃ©trica | PropÃ³sito |
|---------|-----------|
| **IBS** (Integrated Brier Score) | CalibraciÃ³n de probabilidades |
| **Precision/Recall** | Balance de errores Tipo I/II |

### Criterios de Ã‰xito del Experimento

âœ… **Ã‰xito Total**: AlgÃºn escenario aumentado supera baseline en â‰¥5% C-index  
âš ï¸ **Ã‰xito Parcial**: Augmented = Baseline (los sintÃ©ticos no daÃ±an)  
âŒ **Fracaso**: Augmented < Baseline (los sintÃ©ticos degradan el modelo)
